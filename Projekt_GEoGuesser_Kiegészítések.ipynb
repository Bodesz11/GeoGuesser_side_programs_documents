# Diagramok készítése a GeoGuesser projekthez

```python
import pandas as pd
import plotly.graph_objects as go
import plotly.express as px
```

```python
f = open('/content/RESNET50-model_2022-12-30_accuracy.txt','r')

T = {}
i = 0
for x in f:
    x = x.split(':')
    x[1] = float(x[1])
    i+= 1
    T[i] = x

print(T)
```

```python
#Getting the state codes
import pandas as pd
df2 = pd.read_csv('https://raw.githubusercontent.com/plotly/datasets/master/2011_us_ag_exports.csv')

codes=df2['code'].tolist()
print(codes)
```

```python
#Creating the dataframe
df = pd.DataFrame.from_dict(T, orient = 'index')
df.rename({1:'Accuracy'}, axis=1, inplace=True)
df.rename({0:'state_code'}, axis=1, inplace=True)
#df["code"] = ['AL'] + codes.tolist()[:-1]
df["code"] = codes
df
```

```python
#Heatmap of all countries
fig = px.choropleth(df,
                    locations='code', 
                    locationmode="USA-states", 
                    scope="usa",
                    color='Accuracy',
                    #color_continuous_scale="Viridis_r",
                    color_continuous_scale = 'Reds' 
                    
                    )
fig.show()
```

```python
#Heatmap of the worst 5 countries
fig = px.choropleth(df.sort_values(by=['Accuracy']).head(5),
                    locations='code', 
                    locationmode="USA-states", 
                    scope="usa",
                    color='Accuracy',
                    #color_continuous_scale="Viridis_r",
                    color_continuous_scale = 'Blues' 
                    
                    )
fig.show()
```

```python
#Heatmap of the best 5 countries
fig = px.choropleth(df.sort_values(by=['Accuracy']).tail(5),
                    locations='code', 
                    locationmode="USA-states", 
                    scope="usa",
                    color='Accuracy',
                    #color_continuous_scale="Viridis_r",
                    color_continuous_scale = 'Reds' 
                    
                    )
fig.show()
```
